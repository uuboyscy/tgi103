```python
import os
import glob

import jieba


os.listdir("./res_gossiping")
```

```python
glob.glob("./res_gossiping/*.txt")
```

```python
sb = ""

for file_path in glob.glob("./res_gossiping/*.txt"):
    with open(file_path, "r", encoding="utf-8") as f:
        sb += f.read().split("---split---")[0]

sb
```

```python
jieba.load_userdict("./mydict.txt")
word_cut_generator = jieba.cut(sb.replace("\n", " "))
```

```python
word_cut_list = [w for w in word_cut_generator if len(w) > 1]
```

```python
word_cut_list
```

```python
word_count_map = {w: word_cut_list.count(w) for w in word_cut_list}
word_count_map
```

```python
word_count_map = {}

for w in word_cut_list:
    if w in word_count_map:
        word_count_map[w] += 1
    else:
        word_count_map[w] = 1

word_count_map
```
